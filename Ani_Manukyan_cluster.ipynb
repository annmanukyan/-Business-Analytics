```python
import pandas as pd
```

```python
data = pd.read_excel("bike.xlsx",sheetname=3)
```

```python
data.head()
```

```python
from sklearn.cluster import KMeans
```

```python
cluster = KMeans(5)
```

```python
inputs = data.iloc[:,1:]
```

```python
inputs.head()
```

```python
data["cluster"] = cluster.fit_predict(inputs)
```

```python
data.head()
```

```python
data["cluster"].value_counts()
```

```python
new_data = data.iloc[:,[0,-1]]
```
