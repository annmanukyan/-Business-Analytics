```python
import pandas as pd
from statsmodels.formula.api import ols
```

```python
data = pd.read_excel("Conjoint_Survey.xlsx")
data.head()
```

```python
coef_df=pd.DataFrame(columns=range(33))
```

```python
for i in range(33):
    model_ols = ols(formula="Rank ~ A1 + A2 + A3 +B1 + B2 + C1 + C2 + C3", data=data.iloc[i*12:i*12+12,1:])
    our_results = model_ols.fit()
    coef = our_results.params
    coef_df[i]=coef
```

```python
coef_df.head()
```

```python
coef_df.to_excel("Anna_coef.xlsx")
```
